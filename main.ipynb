```python
from dotenv import load_dotenv # pip install python-dotenv
from elevenlabs.client import ElevenLabs # pip install elevenlabs
from IPython.display import Markdown # pip install ipython
import os
```

```python
load_dotenv()

client = ElevenLabs(
    base_url="https://api.elevenlabs.io",
    api_key=os.getenv("ELEVENLABS_API_KEY"),
)
```

```python
# https://elevenlabs.io/docs/models#speech-to-text

MODEL = "scribe_v2"
```

```python
FILE_PATH = './out.mp3'
```

```python
with open(FILE_PATH, "rb") as audio_file:
    response = client.speech_to_text.convert(
        model_id = MODEL,
        file = audio_file,
        language_code = 'en',
        tag_audio_events = False,
        num_speakers = 1,
        timestamps_granularity = 'word', # none, word, character
        diarize = False,
        )
```

```python
Markdown(response.text)
```
